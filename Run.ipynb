```python
def get_weight(x,t):
        alpha,gamma = 0.25,2.
        p = x
        pt = p*t + (1-p)*(1-t)
        w = alpha*t + (1-alpha)*(1-t)
        return w * (1-pt) * gamma
```

```python
import numpy as np
x = np.array([0.9,0.2,0.1])
t = np.array([1,0,0])
rez = get_weight(x,t)
print(rez)
```
